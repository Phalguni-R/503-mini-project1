Qn1: How have housing prices in urban areas changed since 2020?

```python
urban_data = data[data['RegionTypeCheck'] == 'urban']

date_cols_20 = [col for col in data.columns if str(col).startswith('20') and pd.to_datetime(col, errors='coerce') >= pd.to_datetime('2020-01-31')]

date_cols_20 = []
for col in data.columns:
    if '2020' in str(col) or '2021' in str(col) or '2022' in str(col) or '2023' in str(col) or '2024' in str(col) or '2025' in str(col):
        date_cols_20.append(col)

#calculating the statistical analysis
results = []
for _, row in urban_data.iterrows():
    region = row['RegionName']
    prices = [float(row[col]) for col in date_cols]

    results.append({
        'Region': region,
        'Highest': max(prices),
        'Lowest': min(prices),
        'Average': sum(prices) / len(prices),
        'StdDev': np.std(prices)
    })

results_df = pd.DataFrame(results)
results_df = results_df.sort_values('Highest', ascending=False)


```

```python
highest_housing_region = results_df.loc[results_df['Highest'].idxmax()]
lowest_housing_region = results_df.loc[results_df['Lowest'].idxmin()]
most_consistent_region = results_df.loc[results_df['StdDev'].idxmin()]

#printing results
print(f"Highest Housing Value: ${highest_housing_region['Highest']} (Region: {highest_housing_region['Region']})")
print(f"Lowest Housing Value: ${lowest_housing_region['Lowest']} (Region: {lowest_housing_region['Region']})")
print(f"Most Consistent Housing Value: ${most_consistent_region['StdDev']} (Region: {most_consistent_region['Region']})")
```

```python
top5_df = results_df.head(5)

#plotting a bar-chart showing highest housing price for top-5 houses
plt.figure(figsize=(10, 6))
plt.bar(top5_df['Region'], top5_df['Highest'], color='steelblue')
plt.title('Top 5 Urban Regions by Highest Housing Price (2020-2024)')
plt.ylabel('Price ($)')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(top5_df['Highest']):
    plt.text(i, v + 50000, f'${v:,.0f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
#plotting consistency analysis for housing prices
consistency_df = results_df.sort_values('StdDev')

plt.figure(figsize=(10, 6))
plt.bar(consistency_df['Region'].head(5), consistency_df['StdDev'].head(5), color='green', label='Most Consistent')
plt.bar(consistency_df['Region'].tail(5), consistency_df['StdDev'].tail(5), color='orange', label='Least Consistent')
plt.title('Housing Price Consistency by Urban Region (2020-2024)')
plt.ylabel('Standard Deviation ($)')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Qn2: How do urban price trends compare to suburban and rural areas?


```python
urban_data = data[data['RegionTypeCheck'] == 'urban']
suburban_data = data[data['RegionTypeCheck'] == 'suburban']
rural_data = data[data['RegionTypeCheck'] == 'rural']

#date columns
date_cols_1 = [col for col in data.columns if str(col).startswith('20')]
date_cols_1.sort()

urban_avg = []
suburban_avg = []
rural_avg = []

#computing means for urban, rural, suburban
for col in date_cols:
    urban_avg.append(urban_data[col].astype(float).mean())
    suburban_avg.append(suburban_data[col].astype(float).mean())
    rural_avg.append(rural_data[col].astype(float).mean())

trends_df = pd.DataFrame({
    'Date': date_cols,
    'Urban': urban_avg,
    'Suburban': suburban_avg,
    'Rural': rural_avg
})

```

```python
# Simple comparison of first and last prices for each region type
print("\nComparison of Housing Prices:")
for region in ['Urban', 'Suburban', 'Rural']:
    first_price = trends_df[region].iloc[0]
    last_price = trends_df[region].iloc[-1]
    print(f"{region}:")
    print(f"  First recorded price (2000): ${first_price}")
    print(f"  Last recorded price (2024): ${last_price}")
    print(f"  Difference: ${last_price - first_price}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(trends_df['Date'][::12], trends_df['Urban'][::12], label='Urban')
plt.plot(trends_df['Date'][::12], trends_df['Suburban'][::12], label='Suburban')
plt.plot(trends_df['Date'][::12], trends_df['Rural'][::12], label='Rural')

#labels and title
plt.title('Housing Price Trends by Region Type (2000-2024)')
plt.xlabel('Date')
plt.ylabel('Average Price ($)')
plt.legend()

years = [col for col in trends_df['Date'] if col.endswith('-01-31') or col.endswith('-01')]
plt.xticks(years, [year[:4] for year in years], rotation=45)

plt.grid(True, alpha=0.3, linestyle='--', which='major')

plt.tight_layout()
plt.show()
```

Qn3: Which cities have seen the most significant changes in housing prices?


```python
#getting date columns
date_cols_2 = [col for col in data.columns if str(col).startswith('20')]
date_cols_2.sort()

first_date = date_cols[0]
last_date = date_cols[-1]

#calculating percentage changes and absolutes
results = []
for _, row in data.iterrows():
    if pd.notna(row[first_date]) and pd.notna(row[last_date]):
        city = row['RegionName']
        region_type = row['RegionTypeCheck']
        first_price = float(row[first_date])
        last_price = float(row[last_date])

        if first_price > 0:
            abs_change = last_price - first_price
            pct_change = (abs_change / first_price) * 100

            results.append({
                'City': city,
                'RegionType': region_type,
                'First_Price': first_price,
                'Last_Price': last_price,
                'Absolute_Change': abs_change,
                'Percent_Change': pct_change
            })
```

```python
changes = pd.DataFrame(results)

#cities by abs change
abs_change = changes.sort_values('Absolute_Change', ascending=False).head(5)
abs_change = changes.sort_values('Absolute_Change', ascending=True).head(5)

#cities by percentage change
perc_change = changes.sort_values('Percent_Change', ascending=False).head(5)
perc_change = changes.sort_values('Percent_Change', ascending=True).head(5)

```

```python
# Print results
print("Cities with Largest Price Increases (Absolute $):")
print(abs_change[['City', 'RegionType', 'First_Price', 'Last_Price', 'Absolute_Change']])

print("\nCities with Largest Price Increases (Percentage):")
print(perc_change[['City', 'RegionT ype', 'First_Price', 'Last_Price', 'Percent_Change']])
```

```python
#plotting graph
plt.figure(figsize=(12, 8))

#perc change-xaxis, abs change-yaxis
plt.scatter(
    changes['Percent_Change'],
    changes['Absolute_Change'],
    c=['blue' if r == 'urban' else 'orange' if r == 'suburban' else 'green' for r in changes['RegionType']],
    alpha=0.6,
    s=50
)

#labels and title
plt.title('Housing Price Changes by City (2000-2024)')
plt.xlabel('Price Change (%)')
plt.ylabel('Absolute Price Change ($)')
plt.grid(True, linestyle='--', alpha=0.7)

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Urban'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=10, label='Suburban'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Rural')
]
plt.legend(handles=legend_elements)

plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='gray', linestyle='-', alpha=0.3)

plt.tight_layout()
plt.show()
```

Qn4: How did housing prices change during specific time periods (e.g., pre-pandemic vs. post-pandemic)?



```python
#post and pre pandemic dates initialized
pre_pandemic = '2019-03-01'
post_pandemic = '2023-05-01'

#check if dates in columns
date_cols_3 = [col for col in data.columns if str(col).startswith('20')]
date_cols_3.sort()

#if does not exist, go with closest date
if pre_pandemic not in date_cols:
    pre_pandemic = min(date_cols, key=lambda x: abs(pd.to_datetime(x) - pd.to_datetime(pre_pandemic)))
if post_pandemic not in date_cols:
    post_pandemic = min(date_cols, key=lambda x: abs(pd.to_datetime(x) - pd.to_datetime(post_pandemic)))

#calculate by region type
region_types = ['urban', 'suburban', 'rural']
pandemic_changes = []

for region_type in region_types:
    region_data = data[data['RegionTypeCheck'] == region_type]

    #average calculation
    pre_avg = region_data[pre_pandemic].astype(float).mean()
    post_avg = region_data[post_pandemic].astype(float).mean()

    abs_change = post_avg - pre_avg
    pct_change = (abs_change / pre_avg) * 100 if pre_avg > 0 else 0

    pandemic_changes.append({
        'RegionType': region_type,
        'Pre_Pandemic_Avg': pre_avg,
        'Post_Pandemic_Avg': post_avg,
        'Absolute_Change': abs_change,
        'Percent_Change': pct_change
    })

pandemic_df = pd.DataFrame(pandemic_changes)
```

```python
print(f"Housing Price Changes from {pre_pandemic} to {post_pandemic}:")
for _, row in pandemic_df.iterrows():
    print(f"\n{row['RegionType'].capitalize()}:")
    print(f"  Pre-pandemic: ${row['Pre_Pandemic_Avg']}")
    print(f"  Post-pandemic: ${row['Post_Pandemic_Avg']}")
    print(f"  Change: ${row['Absolute_Change']} ({row['Percent_Change']}%)")

```

```python
#plotting a pie-chart showing the distribution between pre and post pandemic
plt.figure(figsize=(20, 8))

#pre-pandemic
plt.subplot(1, 2, 1)
pre_sizes = pandemic_df['Pre_Pandemic_Avg'].values
pre_labels = [f"{r.capitalize()}: ${v:.0f}" for r, v in zip(pandemic_df['RegionType'], pandemic_df['Pre_Pandemic_Avg'])]
colors = ['blue', 'orange', 'green']

plt.pie(pre_sizes, labels=pre_labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.axis('equal')
plt.title(f'Distribution of Housing Prices Pre-Pandemic ({pre_pandemic})')

#post-pandemic
plt.subplot(1, 2, 2)
post_sizes = pandemic_df['Post_Pandemic_Avg'].values
post_labels = [f"{r.capitalize()}: ${v:.0f}" for r, v in zip(pandemic_df['RegionType'], pandemic_df['Post_Pandemic_Avg'])]

plt.pie(post_sizes, labels=post_labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.axis('equal')
plt.title(f'Distribution of Housing Prices Post-Pandemic ({post_pandemic})')

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

#calculate state-level price changes
state_changes = []
for state in data['State'].unique():
    state_data = data[data['State'] == state]
    pre_price = state_data[pre_pandemic].astype(float).mean()
    post_price = state_data[post_pandemic].astype(float).mean()

    change_pct = ((post_price - pre_price) / pre_price * 100) if pre_price > 0 else 0

    state_changes.append({
        'state': state,
        'change_pct': change_pct
    })


state_df = pd.DataFrame(state_changes)

```

```python
#plot choropleth map now
fig = px.choropleth(
    state_df,
    locations='state',
    locationmode='USA-states',
    color='change_pct',
    scope='usa',
    color_continuous_scale='RdYlGn',
    title='Housing Price Changes by State (Pre vs Post Pandemic)',
    labels={'change_pct': 'Price Change (%)'}
)

fig.update_layout(
    geo_scope='usa',
    margin={"r":0,"t":40,"l":0,"b":0}
)

fig.show()
```
